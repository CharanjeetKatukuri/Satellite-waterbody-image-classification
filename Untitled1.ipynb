```python
import zipfile
import os

# Path to your zip file in Google Drive
zip_path = '/content/archive (1).zip'

# Extract the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('/content/waterimages')
```

```python
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_dir = "/content/waterimages"
```

```python
img_size = (224, 224)
batch_size = 32
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    validation_split=0.2
)
```

```python
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='sparse',  # Use 'sparse' for integer labels (0,1,2)
    subset='training'
)
```

```python
val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='sparse',  # Use 'sparse' for integer labels
    subset='validation'
)
```

```python
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(3, activation='softmax')  # 3 categories
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Changed loss function
              metrics=['accuracy'])
```

```python
epochs = 10
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs
)
```

```python
model.save("water_suitability_model.h5")
```

```python
def predict_image(image_path):
    img = keras.preprocessing.image.load_img(image_path, target_size=img_size)
    img_array = keras.preprocessing.image.img_to_array(img) / 255.0
    img_array = tf.expand_dims(img_array, 0)

    predictions = model.predict(img_array)
    class_labels = ["Highly Preferred", "Partially Preferred", "Least Preferred"]

    return class_labels[tf.argmax(predictions[0])]
```
